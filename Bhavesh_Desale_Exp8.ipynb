Q1. What is 2 tailed and  tailed T Test ?

Ans. In hypothesis testing using the **Student's t-test**, the terms **one-tailed** and **two-tailed** refer to **how the alternative hypothesis (H₁) is defined** and **where the rejection region lies**.


1. One-Tailed t-Test

Meaning

A **one-tailed test** checks for a difference in **one specific direction only**.

Types

* **Right-tailed test** → checks if value is **greater than**
* **Left-tailed test** → checks if value is **less than**

Hypothesis

* Right-tailed:

  * H₀: μ = μ₀
  * H₁: μ > μ₀

* Left-tailed:

  * H₀: μ = μ₀
  * H₁: μ < μ₀

Example

* Testing if **average marks are greater than 70**

Rejection Region

* Only **one side (tail)** of distribution


2. Two-Tailed t-Test

Meaning

A **two-tailed test** checks for a difference in **both directions** (either higher or lower).

Hypothesis

* H₀: μ = μ₀
* H₁: μ ≠ μ₀

Example

* Testing if **average marks are different from 70**

Rejection Region

* Both **left and right tails** of distribution


Q2. Explain what is Type I and Type II errors with an example.

Ans. In hypothesis testing (like **Student's t-test** or **Chi-square test**), **Type I and Type II errors** occur when we make **wrong decisions about the null hypothesis (H₀)**.

1. Type I Error (False Positive)

Definition
Rejecting the null hypothesis **when it is actually true**

Symbol

* Denoted by **α (alpha)** → level of significance
Meaning

* You say **“there is an effect”**, but in reality **there is no effect**

Example

Suppose:

* H₀: A medicine has **no effect**

If we conclude:

* “Medicine works” (reject H₀)
* But actually it doesn’t

This is a **Type I Error**



 2. Type II Error (False Negative)

Definition

Accepting the null hypothesis **when it is actually false**

Symbol

* Denoted by **β (beta)**

Meaning

* You say **“no effect”**, but actually **there is an effect*

Example

Same case:

* H₀: Medicine has no effect

 If we conclude:

* “Medicine does not work” (accept H₀)
* But actually it works

This is a **Type II Error**



Q3. What is power of statistical test?

Ans. The **power of a statistical test** is the ability of a test to **correctly detect a real effect** when it actually exists.


Definition

👉 Power is the probability of **rejecting the null hypothesis (H₀) when it is false**

[
\text{Power} = 1 - \beta
]

Where:

* **β (beta)** = probability of **Type II error**



 Meaning

* High power → Test is **good at detecting real differences/effects**
* Low power → Test may **miss real effects**


 Example

Suppose:

* H₀: A new teaching method has no effect

 If the method actually improves performance and:

* The test correctly detects improvement

 This shows **high power**

 If the test fails to detect improvement:
Low power (Type II error)



Factors Affecting Power

1. **Sample Size (n)**

   * Larger sample → Higher power

2. **Effect Size**

   * Bigger difference → Easier to detect → Higher power

3. **Significance Level (α)**

   * Higher α → Higher power (but increases Type I error)

4. **Variability**

   * Less variability → Higher power




Q4.  Explain input and out put characteristics for  stats.ttest_rel function in your own words

Ans. The function **stats.ttest_rel** (from `scipy.stats`) is used to perform a **paired (dependent) t-test**, i.e., when the **same subjects are measured twice** (before–after, matched pairs, etc.).



Input Characteristics

1. Two Related Samples

* Takes **two datasets (arrays/lists)**:

  * `a` → first set of observations
  * `b` → second set of observations

Important:

* Both must be **paired (dependent)**
* Must have **same length**

**Example:**

* Before training scores
* After training scores


 2. Data Type

* Numerical data (**continuous values**)
* Typically floats or integers


 3. Parameters (important ones)

* `a, b` → input samples
* `axis` → direction of computation (default = 0)
* `nan_policy` → how to handle missing values (`propagate`, `omit`, `raise`)
* `alternative` → type of test:

  * `'two-sided'`
  * `'greater'`
  * `'less'`


 Output Characteristics

 Returns Two Values

1. t-statistic

* Measures **difference between paired samples**
* Larger absolute value → stronger evidence against H₀



2. p-value

* Probability of observing the result assuming H₀ is true

 Used for decision:

* If **p-value < α (0.05)**
  ➝ Reject H₀
  ➝ Significant difference exists

* Else
  ➝ No significant difference



 Interpretation

* H₀: Mean difference between pairs = 0
* H₁: Mean difference ≠ 0

 Output tells whether **paired samples differ significantly**



 Example

```python
from scipy import stats

before = [50, 60, 65, 70]
after  = [55, 65, 70, 75]

t_stat, p_val = stats.ttest_rel(before, after)
```

👉 Output:

* `t_stat` → test statistic
* `p_val` → significance level




Q5. What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.

Ans. In Python, the **power of a statistical test** is typically computed using functions from the **statsmodels** (not directly from `scipy.stats`).


Function used to calculate Power

One commonly used function is:

`TTestPower().power()`
From:

```python
from statsmodels.stats.power import TTestPower
```

 Purpose

* Computes the **power of a t-test** given parameters like:

  * Effect size
  * Sample size
  * Significance level (α)



 input Parameters (brief)

* `effect_size` → magnitude of difference
* `nobs` → sample size
* `alpha` → significance level (e.g., 0.05)
* `alternative` → 'two-sided', 'larger', 'smaller'



Output of the Function

 The function returns a **single value: Power**

Output Range

[
0 \leq \text{Power} \leq 1
]

* **0** → No ability to detect effect
* **1** → Perfect detection of effect
